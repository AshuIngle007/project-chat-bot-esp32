```python
#Data loading 
from langchain_community.document_loaders import PyPDFLoader
loader= PyPDFLoader("eval-rhp50000-csl.pdf")
documents=loader.load()
print(documents[0])
```

```python
# data splitting splitting in chunks
from langchain_text_splitters import RecursiveCharacterTextSplitter
textSplitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
finalDocuments = textSplitter.split_documents(documents)
finalDocuments
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model="all-miniLM-L6-v2")
```

```python
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(finalDocuments,embeddings)
db
```

```python
#similarity search 
query = input("Enter your any Query regarding Esp32 :")
docs = db.similarity_search(query)
for i in range(len(docs)):
    print(docs[i].page_content)
```

```python
# Assume you already have docs and query defined
import ollama
context = "\n".join([d.page_content for d in docs[:3]])  # take first few chunks for demo

prompt = f"""
You are an electronics engineering Assistant.
Answer only from the context.
If not found, reply:
"i couldn't find this information in the uploaded document."
Context:
{context}

Question:
{query}
"""

# Send to Ollama
response = ollama.chat(
    model="gemma2:2b",
    messages=[{"role": "user", "content": prompt}]
)

# Print the model's reply
print(response)
```
